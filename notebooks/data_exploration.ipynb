# *Data Exploration*

*Data Exploration* adalah suatu proses uji investigasi awal yang bertujuan untuk mengidentifikasi pola, menemukan anomali, menguji hipotesis dan memeriksa asumsi.

## *Load Library*

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

## *Load Dataset*

```python
# Directory tempat penyimpanan dataset
PATH_DATA_INTERIM='../data/interim/'
```

```python
# load dataset menggunakan pandas
data1 = pd.read_csv(PATH_DATA_INTERIM+'data_interim_clean.csv')
```

```python
data1.sample(10)
```

## *Visualizing Correlation*

Visualisasi dengan *heatmap correaltion matrix* ini digunakan dalam melihat korelasi atau hubungan antar variabel yang bersifat kuantitatif disertai dengan pewarnaan

```python
sns.heatmap(data1.corr(), cmap="ocean")
plt.show()
```

## *Imbalanced Data*

Melihat proporsi jumlah target dengan menggunakan visualisai

```python
sns.countplot(x="level", data=data1, palette="ocean")
plt.show()
```
